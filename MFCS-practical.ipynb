```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Read two images
image1 = cv2.imread('image1.jpg', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('image2.jpg', cv2.IMREAD_GRAYSCALE)

# Display the images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image1, cmap='gray')
plt.title('Image 1')

plt.subplot(1, 2, 2)
plt.imshow(image2, cmap='gray')
plt.title('Image 2')

plt.show()


# Normalize images (scaling pixel values to the range 0 to 1)
image1_normalized = image1 / 255.0
image2_normalized = image2 / 255.0

# Function to calculate statistics
import numpy as np
from scipy import stats

import numpy as np
from scipy import stats

# Function to calculate statistics
def get_image_stats(image):
    try:
        min_val = np.min(image)
        max_val = np.max(image)
        mean_val = np.mean(image)
        std_dev = np.std(image)

        # Get mode and handle cases where mode might be empty
        mode_result = stats.mode(image, axis=None, keepdims=True)  # Ensure compatibility with recent versions of SciPy

        if mode_result.mode.size > 0:
            mode_val = mode_result.mode[0]
        else:
            mode_val = None  # If no mode is found

        return min_val, max_val, mean_val, std_dev, mode_val
    except Exception as e:
        print(f"Error calculating statistics: {e}")
        return None

# Example images (normalized)
image1_normalized = np.random.random((100, 100))  # Example image data
image2_normalized = np.random.random((100, 100))  # Example image data

# Get statistics for both images
stats1 = get_image_stats(image1_normalized)
stats2 = get_image_stats(image2_normalized)

if stats1:
    print(f"Image 1 - Min: {stats1[0]}, Max: {stats1[1]}, Mean: {stats1[2]}, Std Dev: {stats1[3]}, Mode: {stats1[4]}")
if stats2:
    print(f"Image 2 - Min: {stats2[0]}, Max: {stats2[1]}, Mean: {stats2[2]}, Std Dev: {stats2[3]}, Mode: {stats2[4]}")



# Plot histograms for both images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(image1_normalized.ravel(), bins=256, range=[0, 1], color='blue', alpha=0.7)
plt.title('Histogram of Image 1')

plt.subplot(1, 2, 2)
plt.hist(image2_normalized.ravel(), bins=256, range=[0, 1], color='green', alpha=0.7)
plt.title('Histogram of Image 2')

plt.show()

# Combined histogram for both images
plt.figure(figsize=(8, 5))
plt.hist(image1_normalized.ravel(), bins=256, range=[0, 1], color='blue', alpha=0.5, label='Image 1')
plt.hist(image2_normalized.ravel(), bins=256, range=[0, 1], color='green', alpha=0.5, label='Image 2')
plt.title('Combined Histogram of Image 1 and Image 2')
plt.legend()
plt.show()
```
